## 1.1.Web应用框架
> <img style="width:80%; display: inline-block" src="https://resource.xtalker.cn:8000/_static/img/flask-architecture-app.png"/>

### 什么是Web应用框架
        Web应用框架（Web Application Framework）是一种用于开发Web应用程序的软件框架，它提供了一组用于构建Web应用程序的工具、库和约定。使用Web应用框架可以简化Web应用程序的开发过程，提高开发效率，并使得代码更加可维护、可扩展和可重用。

        同时，Python还有TensorFlow、PyTorch等深度学习框架，这些框架提供了丰富的神经网络模型和算法，极大地方便了人工智能应用的开发。
    2.易于学习和使用
        Python语言本身具有简洁明了的语法和易于理解的语义，使得开发者能够更快地掌握和使用。
        基于Python的Web框架，如Django、Flask等，也提供了丰富的功能和工具，帮助开发者快速构建Web应用程序，进一步降低了人工智能应用的开发难度。

## 1.2.Flask的优势
    基于Python的Web应用框架通常有两种选择：
        Django
        Flask
    如下从职业大专学习项目开发的角度分析两者的各自的优势，以及现阶段选择Flask的原因。
    
### Django
    优势：
        强大而全面的功能
        适用于大型项目
    劣势：
        学习曲线较陡峭
### Flask
    优势：
        简单易学，学习曲线平缓
        适合中小型项目
    劣势：
        功能较少

### 总结分析
        Flask 则遵循“微框架”的哲学，只提供必要的核心功能，并鼓励开发者使用扩展（extensions）来添加额外的功能。这并不适合大型项目的开发需求，却非常适合职业大专项目实习的需求。当然最主要的原因还是学习的难度使得Flask成为必然的选择。

## 1.3.Flask基础
    Flask被称为“microframework”，因为它使用简单的核心，通过扩展（extension）来增加其他功能。它的核心组件很少，因此不会给开发者带来过多的负担，相比于大型框架更加灵活和轻便。

### 系统架构
> <img style="width:50%; display: inline-block" src="https://resource.xtalker.cn:8000/_static/img/flask-architecture.png"/>

    Flask主要包括Werkzeug和Jinja2两个核心函数库：
    Werkzeug(WSGI)：
        Werkzeug是一个WSGI（Web Server Gateway Interface）工具集，用于处理HTTP请求和响应；
    Jinja2：
        Jinja2是一个模板引擎，用于渲染HTML页面。

### 使用场景
    Flask适用于各种Web应用程序场景，如个人博客、小型企业网站、RESTful API等。由于其轻量级和可扩展性，Flask在开发者社区中非常受欢迎。
    
### 学习建议
    Flask 学习教程网上资源非常丰富，包括B站和西瓜视频等，有条件可以看油管。
    从开发的角度，初学者重点学习：
        路由映射：
            学习如何将URL请求映射到相应的处理函数（视图函数）。
            理解装饰器在定义URL路由中的作用。
        模板引擎：
            学习如何使用Flask集成的Jinja2模板引擎来渲染动态Web页面。
            掌握模板的继承、包含和宏等高级功能。

# 2.code
    创建自己的项目目录(任何你喜欢的目录名均可，以后就是 Docker-Compose的项目名)：
        flask_base
    在你的项目目录下准备代码文件（文本文件）：
        app.py
        requirements.txt
        Dockerfile
        compose.yml
    系统配置（不是必须）：
        Windows默认位置：C:\Windows\System32\drivers\etc\hosts
        Linux：/etc/hosts

### app.py

```python
# save this as app.py
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello():
    return '''
Hello, World!
[from http://hello.world]
    '''   
```

    应用route()装饰器建立函数对地址(URL)的映射：
        @app.route("/") -- 表示用户访问根目录的响应方法。
    本次课程并不涉及模板的开发，通常的应用都是需要模板来进行支持Web页面的开发。
### requirements.txt

```
flask
```

### Dockerfile

```shell
FROM python:alpine
WORKDIR /code
COPY requirements.txt requirements.txt
RUN pip install -r requirements.txt
CMD ["flask", "run"]
```

### compose.yml

```yaml
services:
  web:
    build: .
    network_mode: bridge
    ports:
     - "80:5000"
    environment:
      - FLASK_APP=app.py
      - FLASK_RUN_HOST=0.0.0.0
    volumes:
      - .\app.py:/code/app.py
```
