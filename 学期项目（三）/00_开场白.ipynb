# 2.项目工作内容
> 学期项目的出发点是项目化教学。那么项目工作的内容是什么？

### 项目化教学
> 项目化教学是一种以学生为中心的教学模式，其核心在于学生实施并完成一个具体的项目过程。在这种教学模式中，教师不再是单纯的知识传授者，而是扮演着指导者和促进者的角色。项目可以是教师提出的，也可以是来自某个企业的真实项目。学生与教师共同讨论，制定学习目标，并在教师的引导下，将项目或任务分解为多个子项目或子目标。

### 项目化学习
> 项目化学习首先是制定一个项目目标，围绕这个目标通过老师指导或自学的方式补充知识短板。

### 项目化教学课堂
> 是一种教学模式，它通过让学生参与实际项目，促进学生对知识的理解和掌握。这种教学方式更加注重学生的实践能力和创新能力。项目化教学以项目为核心，通过学生自主、合作完成具有挑战性的项目来培养学习动力和开拓创新思维。此外，项目化教学的实施步骤包括将班级分成小组，每组课前布置一个项目，如课程项目、毕业设计、实习实训、创新创业等，各组在课前完成方案设计、方案实施、成果分析及项目总结，课堂上各组答辩，教师和学生共同参与项目评价。

## 2.1.从一个想法开始
> 项目最大的特点是有明确的目标。当客户、老板甚至是你你自己有一个想法并希望付诸实施，你就可以启动一个项目。或者同学们项目需要解决的问题难度：线性回归，比如：预测两数相加的结果～直到耗费百亿美元打造的大模型Sora：根据文本生成视频。无论你的项目难度如何几乎都要经理软件开发过程，才能由最初的一个想法变为可以对外提供服务的系统。而在你启动项目直到上线运行过程中遇到的麻烦往往比下面描述的更加复杂和困难，哪怕你只是完成一个 $ 1+1=2 $ 的简单需求。

## 2.2.项目过程（软件为例）
> <img style="width: 100%;display: inline-block" src="img/软件开发流程.jpg">
<br>一个标准化的定制软件系统从项目启动到正式上线的步骤如下：

### 一、需求沟通
> 由项目经理和客户进行详细的沟通，根据客户提出的目标和功能，梳理出适合客户的开发模式，提供给客户一份《产品功能需求逻辑图》，并给出需要的时间和费用。产出物：产品功能需求逻辑图、预估时间，预估费用。

### 二、达成合作
> 在确认功能需求、时间、费用三方面都没有问题之后，签署合同，细节与服务商进行沟通，客户付完首付款项目开发工作正式启动，周期给予客户进度反馈。
产出物：产品功能需求列表确认版，合同

### 三、设计阶段
> 产出物：原型设计文档、UI设计图、数据库设计文档

1. 原型设计
由产品经理设计原型图，采用Axure或者Figma设计工具设计，根据功能需求列表，绘制页面框架线稿图。

2. UI设计
原型图设计完成后，由设计师绘制UI界面，会给出高保真系统界面设计图。

3. 数据库设计
项目技术经理设计根据功能需求列表设计数据库模型，设计表关联关系等。

4. 架构设计
项目技术经理对设计整体的系统分层架构，便于以后再次扩展功能。

### 四、开发阶段
> 全栈开发过程，目前主流的开发方式是，前后端分离开发模式，优点效率高，易扩展。产出物：系统API接口文档

1. 前端开发
前端开发工程师负责根据原型设计图及UI设计图开始编写前端代码，用代码实现设计图中的功能界面，交互，动效等等内容。

2. 后端开发
后端开发工程师负责根据原型设计图、UI设计图，数据库文档编写后端代码，用代码实现业务逻辑接口，如请求参数，请求方法，数据库持久化，数据返回等等内容。

3. 前后端对接
前端和后端人员分别同时开发，互不影响，等到双方都写好功能以后，就可以进行联调了，根据后端开发编写的请求接口文档，前端开始进行代码功能对接，并在其中不断修改和优化接口代码。

### 五、测试阶段
> 测试工程师按阶段设计《测试用例》，未测试通过的需要反馈给开发人员进行调整，开发人员修改代码提交测试，测试通过会在测试结果报告中打钩确认。
产出物：软件测试报告

### 六、软件上线

1. 操作培训
会给用户提供一份详尽的系统操作文档，协助用户使用软件。

2. 试运营
试运营中可能会存在一些问题，可以在正式上线前优化解决掉。

3. 正式上线
系统部署到正式环境，开始正常运行，进入日常维护阶段。

### 七、售后维护
> 提供一定的技术支持。

## 3.2.作为学生，意义何在

### 1.寻求机会
> 通过以上的新闻速览可以看出，人工智能领域巨大的含金量和潜在就业机会。虽然我们没有办法一步到位跨入最先进最具价值的领域从事最尖端的工作，但任何一个巨无霸企业往往都是从一个小项目开始的。比如：OpenAI、Google、微软、TSL等等。通过这门课程了解项目、项目团队合作和其中的技术要求等等，目的是为了让大家在未来能够把握其中的就业机会。因为虽然我们大概率没有机会去做算法、数据分析等高薪领域，但是可能我们有机会去做数据标记、运维、简单的开发，或者承担项目过程中的其它工作。这样我们可以根据自己的优势和特点，在项目运作的众多岗位中寻求工作机会，并选择自己努力的方向。

### 2.补短板
> 之前学习的内容，无论理论知识还是实际操作对于一个完整项目来说存在两方面的问题：1.知识碎片化；2.知识不足。我们需要通过项目将这些碎片化的知识组织起来；并补充不足的部分。

### 3.学期项目课程不是专门的项目管理课程
> 项目管理课程专门课程，不能代替实际项目的演练。本课程会附带讲一些项目管理知识，但不会成体系的讲述项目管理知识。

### 4.学期项目的出发点
> 完成一个项目需要很多的技术支撑、需要多人合作；很多工作难道一定要从零开始吗？项目中的许多工作无法避免，而有些工作却已经无数人走过。难道我们都要重来一遍吗？就像毕昇发明活字印刷术一样。我们学习项目学习的核心是微服务，正是软件领域的活字印刷。要知道大多数（99%以上）软件都是现成的，可以直接拿来使用。我们要做的仅仅拿来用而已。而我们要打开这样一个宝库，往往都是从学习Docker开始。同时微服务还是一个良好项目很作的技术支撑，无论掌握合作技术的人都可以一起工作。比如：有擅长前端界面开发的、有擅长数据管理的和有算法技术的。他们掌握的技能完全不一样，比如：有C语言、Python、JS、PHP等等。微服务可以很好的协助这些人员完成他们各自擅长的工作。
