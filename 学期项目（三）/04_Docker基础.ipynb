# 1.概述


## 1.1.简介
> Docker 是一个开源的应用容器引擎，基于 Go 语言 并遵从 Apache2.0 协议开源。Docker 可以让开发者打包他们的应用以及依赖包到一个轻量级、可移植的容器中，然后发布到任何流行的 Linux 机器上，也可以实现虚拟化。容器是完全使用沙箱机制，相互之间不会有任何接口（类似 iPhone 的 app）,更重要的是容器性能开销极低。Docker 从 17.03 版本之后分为 CE（Community Edition: 社区版） 和 EE（Enterprise Edition: 企业版），我们用社区版就可以了，即Docker-CE。

## 1.2.发展历史
> Docker最初是由dotCloud公司创始人Solomon Hykes在法国期间发起的一个公司内部项目，并于2013年3月以Apache 2.0授权协议开源加入Linux基金会，其项目代码托管于[Github](https://github.com/docker/docker)之上。虽然其最初的实现是基于LXC项目的，但Docker在后来的0.7版本转为使用自行开发的libcontainer容器引擎，而1.11版本又将其换作了runC和containerd。containerd 2017年3月被CNCF接受之后，containerd已经成为行业标准的容器运行引擎。2024年2月28日，云原生计算基金会（CNCF）宣布，containerd是第五个从CNCF毕业的项目。

## 1.3.软件集装箱
> <img style="width: 80%;display: inline-block" src="img/docker集中箱概念.jpg">
<br> &emsp; 英文Docker很好地诠释了Docker集装箱的理念，集装箱可以存放任何货物，可以通过邮轮将货物运输到世界各地。运输集装箱的邮轮和装载卸载集装箱的码头都不用关心集装箱里的货物，这是一种标准的集装和运输方式。
<br> &emsp; 类似的，Docker的容器就是“软件界的集装箱”，它可以安装任意的软件和库文件，做任意的运行环境配置。开发及运维人员在转移和部署应用的时候，不用关心容器里装了什么软件，也不用了解它们是如何配置的。而管理容器的Docker引擎同样不关心容器里的内容，它只要像码头工人一样让这个容器运行起来就可以了，就像所有其他容器那样。
<br> &emsp; 容器不是一个新的概念，但是Docker在对容器进行封装后，与集装箱的概念对应起来了。它之所以被称为“软件界的创新和革命”，是因为它会改变软件的开发、部署形态，降低成本，提高效率。Docker真正把容器推广到了全世界。

## 1.4.虚拟化架构
> <img style="width: 80%;display: inline-block" src="img/docker虚拟化架构.png">
<br> &emsp; Docker并没有传统虚拟化中的Hypervisor层。因为Docker是基于容器技术的轻量级虚拟化，相对于传统的虚拟化技术，省去了Hypervisor层的开销，而且其虚拟化技术是基于内核的Cgroup和Namespace技术，处理逻辑与内核深度融合，所以在很多方面，它的性能与物理机非常接近。

## 1.5.容器文件系统
> <img style="width: 80%;display: inline-block" src="img/docker容器文件系统.webp">
<br> &emsp; Docker的另一个优势是对层级镜像的创新应用，即不同的容器可以共享底层的只读镜像，通过写入自己特有的内容后添加新的镜像层，新增的镜像层和下层的镜像一起又可以作为基础镜像被更上层的镜像使用。
<br> &emsp; 这种特性可以极大地提高磁盘利用率，所以当你的系统上有10个大小为1GB的镜像时，它们总共占用的空间大小可能只有5GB，甚至更少。
<br> &emsp; 另外，多个容器使用同一个基础镜像时，可极大地减少内存占用。

## 1.6.系统架构
> <img style="width: 80%;display: inline-block" src="img/docker-architecture.webp">
<br> Docker的核心组件包括Docker客户端、Docker守护进程、Docker镜像、容器和仓库等。由这些核心组件构成Docker系统架构。

### 1.Docker客户端
&emsp; 是与Docker守护进程交互的主要方式，可以通过命令行界面（CLI）执行。

### 2.Docker daemon
&emsp; 运行于主机上的长时间进程，用于处理Docker API请求，管理Docker对象。

### 3.Docker容器
&emsp; 是镜像的运行实例，可以运行、开始、停止、移动等。

### 4.Docker镜像
&emsp; 是创建Docker容器的模板，可以包含文件系统和其他配置信息。

### 5.Registry
&emsp; Registry是一个存放镜像的中心仓库，它通常被部署在互联网服务器或者云端。Registry国内经常叫做镜像仓库，其实是有一些区别，正式叫法应该是注册服务器。除此之外网络上还经常提到名词`镜像源`其实是镜像缓存服务器，主要是解决访问速度慢的问题，不过注意的是这些镜像缓存服务器大多数不可用，应尽量避免使用。

## 1.7.相关概念

### 1.DockerEngine
    Docker典型的客户端/服务器模式来运行的，包括上述的Docker客户端和Docker守护进程（DockerDaemon）。Docker daemon也可以被理解成Docker Server，也常常用Docker Engine来直接描述它，因为它就是驱动整个Docker功能的核心引擎。
    比较特别的地方在于，Linux系统的Docker发布版本，Docker将客户端和服务器端统一在同一个二进制文件中。正式基于这种Docker客户端和守护进程之间密不可分的状况，很多情况下说起DockerEngine往往会包括Docker客户端。不过，这只是对于Linux系统而言的，在其他平台如Mac上，Docker单独提供了客户端。
    另外Docker Engine这种叫法也是说明它是整个Docker软件平台的核心部分，

### 2.Compose
    Compose 项目是 Docker 官方的开源项目，负责实现对 Docker 容器集群的快速编排。通过 Compose 可以方便的定义和运行多容器 Docker 应用程序；通过 Compose，您可以使用 YML 文件来配置应用程序需要的所有服务。然后，使用一个命令，就可以从 YML 文件配置中创建并启动所有服务。

### 3.Kubernetes
    Kubernetes 这个名字源于希腊语，意为“舵手”或“飞行员”，简称 k8s，8代表中间的八个字符，总共10个字符。Google 在 2014 年开源了 Kubernetes 项目，是Google基于Borg开源的容器编排调度引擎。作为CNCF（Cloud Native Computing Foundation）最重要的组件之一，它不仅仅是一个编排系统，而是一种规范或者标准，目前已是容器生态的标准。

## 2.4.Registry（注册服务器）
> <img style="width: 80%;display: inline-block" src="img/Docker系统架构图.png">
<br> 注册服务器如上图所示的右边部分，分为公共和私有两种类型：
<br> &emsp; 对于公共服务器来说，无需安装也无需注册，只要网络可达即可使用。但是服务提供商可能会对下载内容和下载速度有一定的限制。
<br> &emsp; 对于多数企业和研究机构通常会选择自己部署Docker Registry或者购买功能更强大的企业版Docker Hub。这样就可以避免限速和互联网传输风险等问题。

### 全球公共注册服务器
公共镜像仓库 | 说明
:-- | :--
Docker Hub | 这是 Docker 官方的公共镜像仓库，也是全球最大的镜像市场。它允许用户注册账号，上传和分享自己的镜像，同时也提供了大量的官方镜像和社区维护的镜像。
Google Container Registry (GCR) | Google 提供的容器镜像仓库服务，与 Google Cloud Platform (GCP) 集成紧密。用户可以在 GCP 中创建和管理自己的镜像仓库，并用于存储和分发 Docker 镜像。
Amazon Elastic Container Registry (ECR) | Amazon 提供的容器镜像仓库服务，与 Amazon Web Services (AWS) 集成。用户可以在 AWS 中创建和管理自己的镜像仓库，用于存储和分发 Docker 镜像。
Quay.io | 这是一个由 Red Hat 提供的公共镜像仓库，它提供了与 Docker Hub 类似的功能，但也有一些额外的特性，如更细粒度的访问控制和镜像签名等。
Azure Container Registry (ACR) | Microsoft 提供的容器镜像仓库服务，与 Azure 集成。用户可以在 Azure 中创建和管理自己的镜像仓库，用于存储和分发 Docker 镜像。
docker.pkg.github.com | GitHub 也提供了 Docker 镜像仓库的服务。每个 GitHub 用户都有一个与之关联的 Docker 镜像仓库，可以通过 `docker.pkg.github.com/<username>/<repository>` 的地址进行访问。

### 注意事项

* 公共注册服务器：普通开发者一般只需要用到公共镜像仓库`Docker Hub`就足够了
* 私有注册服务器：如果你是在企业学校等机构可能会因为信息安全、网络带宽等原因可能需要搭建和使用私有注册服务器。
* 镜像缓存服务器：网络上有很多国内镜像源的教程，其实正式叫法是缓存服务器，目前网络上介绍的大多数缓存服务器基本都已经废弃不可用或者不对外提供服务。如果确实需求应该在你的组织内部自建。

### 2.`search`命令（没有办法的办法）
> &emsp; 不过 `hub.docker.com` 网址在国内无法直接使用，需要进行科学上网。包括开发人员一辈子绕不过的 `github.com` 都是需要有一些上网手段的。除此之外还可以使用命令行的方式来找到镜像资源，这个不受防火墙限制，只是信息量要少很多。

````shell
# 1. 帮助
docker search --help

# 2. 搜索数据库mysql镜像
docker search mysql

# 3. 限制结果集
docker search --limit=5 mariadb

# 4. 过滤器的用法
# 过滤非官方镜像
docker search --filter is-official=true busybox

#  过滤推荐星级，星数大于10
docker search -f stars=10 mariadb

# 5. 多种过滤条件组合
docker search busybox --filter is-automated=true --filter stars=3

# 6. 格式化输出字段
docker search --format "{{.Name}} \t {{.StarCount}}" nginx
# 表格，表头方式输出
docker search --format "table {{.Name}} \t {{.StarCount}}" nginx
docker search --format "table {{.Name}}\t{{.IsAutomated}}\t{{.IsOfficial}}" nginx
````

## 3.2.从一个流行操作系统开始
> 快速学习的过程中遇到所有指令都可以通过命令行帮助、书籍和网络资源找到答案，这里不介绍指令的基本使用。
<br> &emsp; 我们的任务是基于乌班图 `ubuntu` 创建一个我们自己的镜像，并尝试使用它。如果你连大名鼎鼎的乌班图是什么都不知道，可以访问一下官网：https://cn.ubuntu.com/ 。

### 1.下载镜像

````shell
# 命令不清楚问帮助
docker --help

# 找一下
docker search ubuntu

# 下载
docker pull ubuntu

# 查询镜像信息
docker scout quickview ubuntu
````
### 2.启动实例
> 与之前的Hello world镜像的使用不同，`ubuntu` 作为操作系统需要一直在后台运行，并且需要提供一个可以供我们交互式命令行方式操作的界面。因此我们需要学习到两个重要参数：`-it` 和 `-d`。具体解释问帮助。

```shell
# 启动容器实例，我们给容器取一个名字，便于使用它。如果不指定，Docker会随机指定一个名字。
docker run -it -d --name=ubuntu ubuntu

# 由于使用 -d 参数，启动容器后会直接返回用户界面。我们需要查询这个进入后台运行的容器实例。
docker container ls

# 如果以上命令没有找到你启动容器，你给它取的名字叫做 ubuntu。那么可能这个容器死掉了，你需要使用如下命令来查询。因为默认查询只查询目前启动状态的容器，死掉的容器需要带 -a 参数。
docker container ls -a

```

### 3.使用

````shell
# 进入 ubuntu 容器的虚拟环境，你就像拥有一台 独立的Linux操作系统一样使用它。
docker exec -it ubuntu bash
# 执行要给命令，看看输入是否一致？
cat /etc/os*
````

## 3.3.Ubuntu的使用
> Ubuntu的使用并不是Docker的学习内容，但对于一个项目的实现，Linux的知识点太重要了，你可以尝试和它熟悉熟悉打个招呼。

````shell
# 进入容器内部，之下的指令均是进入容器内部执行的
docker exec -it ubuntu bash

# 安装认证中心（apt可以换成apt-get，看个人习惯），安装好慢，因为服务器在国外。
apt update
apt install ca-certificates apt-transport-https apt-utils

# 国外安装源太慢，换一下。首先备份一下配置文件是个好习惯
cp -f /etc/apt/sources.list /etc/apt/sources.list.bak
sed -i "s@http://.*archive.ubuntu.com@https://mirrors.tuna.tsinghua.edu.cn@g" /etc/apt/sources.list
sed -i "s@http://.*security.ubuntu.com@https://mirrors.tuna.tsinghua.edu.cn@g" /etc/apt/sources.list

# 现在更新安装源是否快很多呢?如果有错误需要仔细检查。
apt update

# 设置中文运行环境
apt install locales language-pack-zh-hans
sed -i -e "s/# en_US.UTF-8 UTF-8/en_US.UTF-8 UTF-8/" /etc/locale.gen
sed -i -e "s/# zh_CN.UTF-8 UTF-8/zh_CN.UTF-8 UTF-8/" /etc/locale.gen
update-locale LANG=zh_CN.UTF-8
dpkg-reconfigure --frontend=noninteractive locales

# 退出
exit
# 重启容器（不是必须）
docker restart ubuntu
# 重新进入中文环境
docker exec -it ubuntu bash -c "LANG='zh_CN.UTF-8' LC_ALL='zh_CN.UTF-8' /bin/bash"

# 测试验证一下，看看是不是中文提示了呢？
apt install fortunes-zh

# 运行一下刚才安装的软件：fortunes
/usr/games/fortune
````
