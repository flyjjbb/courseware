# 1.概述
## 1.1.简介
        Ubuntu是一个以桌面应用为主的Linux发行版操作系统，它的名称源自非洲南部祖鲁语或豪萨语的“ubuntu”一词，意为“人性”和“我的存在是因为大家的存在”，这体现了非洲传统的一种价值观。
        Ubuntu基于Debian发行版和GNOME桌面环境，其设计原则为简洁、直观和易用，适合各类用户。它提供了友好的图形界面，支持多种语言和自定义设置，用户可以根据自己的喜好和需求进行个性化设置。同时，Ubuntu注重安全性和稳定性，定期发布更新和修补程序，以保护用户的系统安全。
        Ubuntu的发展历史可以追溯到2004年，其首个正式版本Ubuntu 4.10 Warty Warthog于当年10月发布。此后，Ubuntu不断发展和完善，逐渐在Linux发行版领域占据领先地位。
        此外，Ubuntu还拥有庞大的社区力量，用户可以方便地从社区获得帮助。这使得Ubuntu成为开源和易用的Linux操作系统的代表之一。无论是初学者还是高级用户，都能在Ubuntu社区中找到所需的资源和支持。

## 1.2.版本

### 长期支持（LTS）版本
    LTS版本是Ubuntu为那些需要更长时间稳定支持和较少升级的用户提供的。
    LTS版本每两年发布一次，并在发布后的五年内提供安全更新和维护。
    LTS版本通常都是偶数版本。
    与之对应的是最新开发版本，通常是奇数版本，我们通常选择偶数的LTS版本。

### 版本命名
    Ubuntu的每个版本都有一个独特的代号，这些代号通常与动物有关，并且按照字母顺序递增。
        Ubuntu 16.04的代号：“Xenial Xerus”
        Ubuntu 18.04的代号是“Bionic Beaver”
        Ubuntu 20.04的代号是“Focal Fossa”
        Ubuntu 22.04的代号是“Jammy Jellyfish”
    你可以通过如下命令在Ubuntu系统中查看当前的版本和代号：
```shell
lsb_release -a
No LSB modules are available.
Distributor ID: Ubuntu
Description:    Ubuntu 22.04.4 LTS
Release:        22.04
Codename:       jammy

cat os-release | grep VERSION
VERSION_ID="22.04"
VERSION="22.04.4 LTS (Jammy Jellyfish)"
VERSION_CODENAME=jammy
```

### 版本选择
    截至2024年4月，Docker-hub上的Ubuntu默认版本号为：22.04，即当前主流LTS版本。因此如下的镜像其实是同一个：
        docker pull ubuntu
        docker pull ubuntu:22.04
    注意：
        Ubuntu的版本更新迭代非常快，小版本的更新是相互兼容的。而大版本更新就不一定了。因此强烈建议使用第二种镜像下载方式。

## 3.2.自动构建-V2.0

### Dockerfile
    创建一个文本文件：“Dockerfile”，注意没有扩展名，其内容如下：
```shell
# docker build --tag local/ubuntu:2.0 .
FROM ubuntu:22.04

WORKDIR /root
RUN apt update;apt install -y curl net-tools procps openssh-server nginx

ENTRYPOINT ["bash", "-c"]
CMD ["service ssh restart;service nginx restart;bash"]

```

### build指令

```shell
# 你需要进入 Dockerfile 文件所在的目录
docker  build --tag local/ubuntu:2.0 .
...

docker images
---------------------------
```
