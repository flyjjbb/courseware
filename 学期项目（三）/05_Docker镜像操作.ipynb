# 1.`login&push`

## 1.1.注册服务器

### 1.登录&注册

* 私有注册服务器地址
> https://docker.xtalker.cn:8443

* 注册和登录过程略

### 2.创建或者加入项目
> 匿名用户是没有写入权限的，需要浏览器登录注册服务器，自建项目或者加入已有项目。

### 3.Docker登录

```shell
docker login -u [账号] docker.xtalker.cn:8443

Password: 输入账号密码

Login Succeeded
```
