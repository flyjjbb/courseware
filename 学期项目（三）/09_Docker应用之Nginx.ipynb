# Docker应用：Nginx
        Nginx（发音为“engine-x”），是一个高性能的HTTP和反向代理服务器，也是一个IMAP/POP3/SMTP代理服务器。Nginx的历史可以追溯到2002年，当时由俄罗斯程序员伊戈尔·赛索耶夫为俄罗斯访问量第二的Rambler.ru站点（一家俄罗斯搜索引擎和门户网站）开发了一个反向代理服务器。随着需求的增加，Igor Sysoev逐渐扩展了这个项目，并于2004年发布了最初版本的Nginx。
        Nginx开发的初衷是为了解决C10K问题（即单台服务器支持超过1万个并发连接）而设计的。在早期的Web服务器和反向代理服务器中，处理大量并发连接通常需要大量的系统资源，而Nginx的设计目标就是在资源有限的情况下，实现高并发、高性能的Web服务。
    本章内容：学习Nginx官方镜像

### 相关项目
> [<img style="display: inline-block;zoom:50%;" src="https://resource.xtalker.cn:8000/_static/img/apahce-1999-logo.jpg" />](https://httpd.apache.org/)

        Apache HTTP Server（通常简称为Apache httpd或httpd）是Apache软件基金会（ASF）的一个开源的网页服务器，可以在多种操作系统上运行，包括Unix、Linux、Windows等。它是全球使用最广泛的Web服务器软件之一，以其稳定性、安全性、可定制性和灵活性而著称。
        Apache HTTP Server的名字来源于其开发者对Apache印第安人的纪念，因为Apache族是最后一个屈服于美国政府的民族，被认为具有高超的作战策略和无穷的耐性。同时，Apache这个名字也被解释为“一个修补的（a patchy）”服务器，因为Apache HTTP Server最初是由NCSA HTTPd服务器的代码修改而成的。

### Apache | Nginx
        在选择使用Apache还是Nginx时，需要根据具体的场景和需求进行考虑。如果需要一个功能全面、支持多种动态脚本语言的Web服务器，并且处理动态请求较多，那么Apache可能是一个更好的选择。而如果需要一个轻量级、高性能、擅长处理静态内容和反向代理的Web服务器，并且处理高并发请求较多，那么Nginx可能更适合。

# 1.概述
## 1.1.Nginx组网图
> <img style="zoom:50%; display: inline-block" src="https://resource.xtalker.cn:8000/_static/img/nginx组网图.png"/>

    Nginx功能特点（注意不仅限于如下的功能点）：
        1.web服务器
            静态页面的web服务器；
            动态页面的web服务器：支持CGI协议的动态语言，如perl、php等；
            动态页面的web服务器：通过WSGI（Web Server Gateway Interface）实现对Python语言的支持；
            动态静态页面：将动态页面（如jsp）和静态资源（如html、css、jpg）分别放在不同的服务器上，这样Nginx可以根据不同类型的资源请求，将请求转发到相应的服务器，从而提高网站的解析速度并降低单个服务器的压力。
        2.反向代理
            客户端发送的请求首先到达Nginx服务器，然后由Nginx服务器转发到真正的服务器（如tomcat）。
            这种机制可以隐藏真实服务器的地址，只暴露Nginx服务器的地址。
        3.负载均衡
            为了减少单个服务器的访问压力，可以设置多个服务器来分担负载。
            Nginx可以根据一定的策略（如轮询、ip_hash、权重）来决定将请求转发到哪一个服务器。
        4.热部署
            在不停止服务的情况下，可以更新或替换Nginx的配置文件或某些模块，这对于维护和升级非常有用。
        5.高性能高并发
            Nginx的最大并发连接数可以达到50000个，这使得它非常适合处理高并发的网络请求。

### 挂载过程

```shell
docker stop nginx

cd C:\Btsync\teacher\docker
docker run --rm -it -d --name nginx --hostname nginx -p 80:80 -v .\sources.list:/etc/apt/sources.list nginx:stable
```
> 挂载的方式 宿主机目录/文件:容器目录/文件

    宿主机目录/文件写法：
        1.挂载当前目录
            Windows：
                .\file
                .\subdir\file
            Linux:
                ./file
                ./subdir/file
            不能直接使用目录/文件名：因为这是数据卷模式
        2.挂载绝对路径
            Windows：
                C:\Btsync\teacher\docker\sources.list
            Linux：
                /Btsync/teacher/docker/sources.list
            不能直接使用目录名开头：因为这是数据卷模式
    注意事项：
        宿主机目录不存在，Docker自动创建（仅限一层目录）
        容器目录不存在，Docker自动创建（仅限一层目录）
        容器中的文件已经存在，映射文件会覆盖它。（注意不是删除，回忆一下联合文件系统）
